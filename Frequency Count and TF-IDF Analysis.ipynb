```python
import pandas as pd
import numpy as np
import re
import nltk
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Download stopwords if not already downloaded
nltk.download("stopwords")
nltk.download("punkt")
```

```python
df_fiction = pd.read_csv("df_fiction.csv")
df_fiction.head()
```

```python
# display the number of nan in each column
df_fiction.isnull().sum()
```

```python
reviews = df_fiction["review/text"].dropna().tolist()  # Drop missing reviews
```

```python
# Function for text preprocessing
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"[^\w\s]", "", text)  # Remove punctuation
    words = word_tokenize(text)  # Tokenize text
    words = [word for word in words if word not in stopwords.words("english")]  # Remove stopwords
    return words
```

```python
vectorizer = TfidfVectorizer(stop_words="english", max_features=20000)  # Built-in stopword removal
tfidf_matrix = vectorizer.fit_transform(reviews[:10000])  # Run on a sample first
```

```python
# Get feature names (words) and their importance scores
tfidf_words = vectorizer.get_feature_names_out()
tfidf_scores = np.asarray(tfidf_matrix.mean(axis=0)).flatten()  # Compute average importance
```

```python
# Sort words by importance
tfidf_results = sorted(zip(tfidf_words, tfidf_scores), key=lambda x: x[1], reverse=True)
```

```python
print("\nðŸ”¹ Most Important Words by TF-IDF:")
for word, score in tfidf_results:
    print(f"{word}: {score:.4f}")
```
